# Conclusão 

§ De acordo com os resultados calculados anteriormente e com a condição histórica conclui-se claramente que o
melhor que todo cidadão brasileiro possuinte de recursos financeiros disponíveis em sua carteira é investir em ativos,
ou mesmo comprar dólar,nunca deixar em conta parado, sem aplicação, ou mesmo em poupança(diminui a perda somente).
Para título de curiosidade a poupança rendeu neste período de 12 anos --> 67.33% ou seja, nem cobriu a inflação de 73.57% 

O resultado demonstra o quão a inflação é corrosiva ao poder de compra em curto e longo período de tempo no Brasil.

§ Apesar do Bitcoin ser uma tecnologia disruptiva na área econômica e financeira os fundamentos do mesmo reflete ao longo
prazo até o momento na valorização do ativo, mesmo com queda substancial recente demonstrou uma valorização surpreendente e rápida em torno de 10 anos, sendo o pico máximo entorno de 68 mil dólares, fato a ser estudado e considerado com a devida cautela.

§ Devido a gigantesca dimensão do Brasil, os imóveis mesmo com a disparidade entre as regiões demonstrou-se passíveis de 
serem considerados a sua aquisição e manutenção que seja alugando ou mesmo revendendo no futuro, porém de acordo com a vivência do autor em sua cidade rio verde - go, os imóveis na região sofreram uma gigantesca valorização frente 
o índice nacional, muito superior a média nacional, cada região há a sua oferta e demanda.

§ O índice ibovespa é composto por uma carteira teórica de ações, pelos papéis com maior volume financeiro da Bolsa de Valores do Brasil por um determinado período, o Ibovespa ajuda bastante na hora de fazer o acompanhamento da rentabilidade média dessas ações importantes, movedoras da economia brasileira. Todavia o índice não demonstra tão bem a realidade das empresas componentes desse índice, sendo que, com pouca pesquisa encontra-se empresas de boa gestão e de setores perenes que se valorizaram mais que o dobro do índice no mesmo período, havendo também aquelas que o papel sofreu desvalorização. As ações por ser renda variável deve ser conduzida com cautela, com muita pesquisa e dedicação sobre a empresa a ser investida, inclusive sobre o corpo administrativo, todavia o índice ibovespa demonstrou a média ser superior a inflação do período, sem contar os proventos(dividendos e JCP) recebidos por possuir ações.  



## Conclui-se:

Com as condutas políticas adotadas pelos governantes há um empobrecimento geral do brasileiro na última década que mantém dinheiro disponível em conta mesmo que seja poupança, principalmente ao não possuidor de ativos, contudo essa singela demonstração prova a necessidade do brasileiro mesmo de baixa renda procurar ativos pois no tempo demonstrou-se a sua rentabilidade ser superior ao mínimo que é a inflação.
